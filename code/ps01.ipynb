```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
```

```python
wave = thinkdsp.read_wave('Analog23.wav')
wave.make_audio()
```

Analog23.wav sourced from http://www.freesound.org/people/junggle/sounds/31246/

```python
wave.plot()
```

```python
start = 1.1
duration = 0.5
segment = wave.segment(start, duration)
segment.plot()
```

2) The majority of the wave does look sinusodial.

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

3) There are no harmonic structures that I can determine.

```python
segment.make_audio()
```

# High Pass Filter

```python
spectrum = segment.make_spectrum()
spectrum.high_pass(2000)
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
filtered.plot()
```

```python
filtered.make_audio()
```

# Low Pass Filter

```python
spectrum = segment.make_spectrum()
spectrum.low_pass(1000)
spectrum.plot()
```

```python
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
filtered.plot()
```

```python
filtered.make_audio()
```
